```python
def AnalyzeFrance():
  matchesplayed = df['home_team'].value_counts()['France']
  print(f'Total number of matches played by France is: {matchesplayed}')
```

```python
def AnalyzeTown():
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    print(f'Total number of matches played by (_) is : {TownData[_]}')
```

```python
AnalyzeTown()
```

```python
def MatchesPlayedByTeam():
    townName = []
    matchesPlayed = []
    TownData = df['home_team'].value_counts()
    for _ in TownData.index:
      if TownData[_] > 30 :
        townName.append(_)
        matchesPlayed.append(TownData[_])
    plt.bar(matchesPlayed,townName)
    plt.xlabel('Town Name')
    plt.ylabel('Matches Playesd')
    plt.title('Matches Played by Town')
    plt.show()
```

```python
MatchesPlayedByTeam()
```

```python
#Function to get a graph
def getGraph(DataFrame,Column1,Column2):
  from matplotlib import pyplot as plt

  #Check the length of column1 and column2
  if len(DataFrame[Column1]) == len(DataFrame[Column2]):
    df.plot(x= Column1,y = Column2, kind= 'bar',color = ['Yellow','Red'])
    print("The Graph/Plot is successfully created")

  else:
      print("Please check the column size once")
```
