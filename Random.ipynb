```python
import random
```

```python
a1 = []
a2 = []
a3 = []
a4 = []
a5 = []
```

```python
for _ in range(1000):
    a1.append(random.randint(0,100))
    a2.append(random.randint(50,70))
    a3.append(random.randint(10,20))
    a4.append(random.randint(5,10))
    a5.append(random.randint(20,80))
```

```python
df = {
    'grade': a1,
    'height': a2,
    'friends': a3,
    'cousins' : a4,
    'projects' : a5
    }
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(df)
```
