```python
#import necessary libraries
import pandas as pd
```

```python
datasetpath = '/content/drive/MyDrive/retail_transactions.csv'
```

```python
#Data loading
#Read the CSV file into a pandas DataFrame
df = pd.read_csv(datasetpath)
```

```python
# 2. Initial Data Exploration
# Display the first 15 rows of the DataFrame
print(df.head(15))
```

```python
# Print a summary of the DataFrame
print(df.info())
```

```python
print(df.describe())
```

```python
# Data Cleaning
# Check for missing values in each column and print the count per column
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
# Ensure Quantity, Price, and Discount are in the correct numerical format
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')
```

```python
df
```

```python
# Fill missing Discount values with 0%
df['Discount'].fillna(0, inplace=True)
```

```python
# Feature Engineering
# Create a new column called NetPrice
df['NetPrice'] = df['Price'] * (1 - df['Discount'] / 100)
```

```python
# Create another new column called TotalSale
df['TotalSale'] = df['Quantity'] * df['NetPrice']
```

```python
#Check the newly created columns
df.columns
```

```python
# Data Filtering
# Filter the DataFrame where TotalSale is >= 20
filtered_df = df[df['TotalSale'] >= 20]
```

```python
filtered_df
```

```python
# Data Type Conversion
# Convert the PaymentType column to a categorical data type
filtered_df['PaymentType'] = filtered_df['PaymentType'].astype('category')
```

```python
print(type(filtered_df['PaymentType'].dtype))
```

```python
# Grouping and Aggregation
# Group the filtered DataFrame by ProductCategory and compute total Quantity and TotalSale
grouped_category = filtered_df.groupby('ProductCategory').agg(
    TotalQuantity=('Quantity', 'sum'),
        TotalSales=('TotalSale', 'sum')
        ).reset_index()
```

```python
grouped_category
```

```python
# Sort the grouped result in descending order based on the total TotalSale
sorted_grouped_category = grouped_category.sort_values(by='TotalSales', ascending=False)
```

```python
sorted_grouped_category
```

```python
# Identify and print the top 3 product categories with the highest total sales
top_3_categories = sorted_grouped_category.head(3)
print("Top 3 Product Categories with Highest Total Sales:\n", top_3_categories)
```

```python
# Multi-Level Grouping
# Group the filtered data by both StoreID and ProductCategory
multi_grouped = filtered_df.groupby(['StoreID', 'ProductCategory']).agg(
    AverageTotalSale=('TotalSale', 'mean')
    ).reset_index()
```

```python
multi_grouped
```

```python
# Store-Level Insights
# For each store (StoreID), determine which product category has the highest average TotalSale
store_insights = multi_grouped.loc[multi_grouped.groupby('StoreID')['AverageTotalSale'].idxmax()]
```

```python
print(store_insights)
```

```python
# Prepare and print a summary DataFrame
summary_df = store_insights[['StoreID', 'ProductCategory', 'AverageTotalSale']]
print("Store-Level Insights:\n", summary_df)
```
