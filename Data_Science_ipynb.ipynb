```python
def AnalyzeFrance():
  matchesplayed = df['home_team'].value_counts()['France']
  print(f'Total number of matches played by France is: {matchesplayed}')
```

```python
AnalyzeFrance()
```

```python
def AnalyzeTown():
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    print(f'Total number of matches played by (_) is : {TownData[_]}')
```

```python
AnalyzeTown()
```

```python
%matplotlib inline
```

```python
def MatchesPlayedByTeam():
    townName = []
    matchesPlayed = []
    TownData = df['home_team'].value_counts()
    for _ in TownData.index:
      if TownData[_] > 30 :
        townName.append(_)
        matchesPlayed.append(TownData[_])
    plt.bar(matchesPlayed,townName)
    plt.xlabel('Town Name')
    plt.ylabel('Matches Playesd')
    plt.title('Matches Played by Town')
    plt.show()
```

```python
MatchesPlayedByTeam()
```

```python
from matplotlib import pyplot as plt
```
